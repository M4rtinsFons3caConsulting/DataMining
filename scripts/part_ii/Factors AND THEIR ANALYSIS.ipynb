# NMF

```python
from sklearn.decomposition import NMF
from sklearn.preprocessing import MinMaxScaler

# Preprocess the data (Standardize it)
scaler = MinMaxScaler()
data_scaled = scaler.fit_transform(data)
```

```python
# Apply NMF to extract latent factors
nmf = NMF(n_components=5, init='random', random_state=0, solver='mu', beta_loss='kullback-leibler')
W = nmf.fit_transform(data_scaled).round(decimals=3) # Customer-factor matrix
H = nmf.components_.round(decimals=3)  # Cuisine-factor matrix

# Create a DataFrame for W, and H with appropriate row and column names
H_df = pd.DataFrame(H, columns=cuisines, index=[f"Factor_{i+1}" for i in range(H.shape[0])])
W_df = pd.DataFrame(W, columns=[f"Factor_{i+1}" for i in range(W.shape[1])], index=data.index)
```

```python
# Top labels for H (cuisine-factor matrix)
top_labels_H = {}
for factor in H_df.index:
    # Sort cuisines for the current factor
    sorted_cuisines = H_df.loc[factor].sort_values(ascending=False)
    top_cuisines = sorted_cuisines.head(3).index.tolist()  # Top 3 cuisines
    top_labels_H[factor] = {"Top Cuisines": top_cuisines}

# Display top cuisines for each factor in H
print("Top Cuisines for Each Factor (H):")
for factor, labels in top_labels_H.items():
    print(f"{factor}:")
    print(f"  Top Cuisines: {', '.join(labels['Top Cuisines'])}")
    print()

# Top labels for W (customer-factor matrix)
top_labels_W = {}
for factor in W_df.columns:
    # Sort customers for the current factor
    sorted_customers = W_df[factor].sort_values(ascending=False)
    top_customers = sorted_customers.head(3).index.tolist()  # Top 3 customers
    top_labels_W[factor] = {"Top Customers": top_customers}
```

# sPCA

```python
import numpy as np
import pandas as pd
from sklearn.decomposition import SparsePCA
from sklearn.preprocessing import StandardScaler

# Assuming you have your dataset stored in 'data'
# Preprocess the data (Standardize it)
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data)

# Apply Sparse PCA to extract sparse components
n_components = 5  # Number of components you want to extract
sparse_pca = SparsePCA(n_components=n_components, alpha=0.5, random_state=0)

# Fit the Sparse PCA model
sparse_pca.fit(data_scaled)

# Get the Sparse PCA components
components = sparse_pca.components_

# Get the top labels for Sparse PCA components (top values in each component)
top_labels = {}
for i, component in enumerate(components):
    sorted_indices = np.argsort(component)[::-1]  # Sort in descending order
    top_features = data.columns[sorted_indices][:3]  # Top 3 features
    top_labels[f'Component_{i+1}'] = top_features

# Display top labels for each Sparse PCA component
print("\nTop Labels for Each Sparse PCA Component:")
for component, labels in top_labels.items():
    print(f"{component}: {', '.join(labels)}")
    
# Convert the components into a DataFrame with cuisines as columns and components as rows
components_df = pd.DataFrame(components, columns=data.columns, index=[f"Component_{i+1}" for i in range(components.shape[0])])

# Display the DataFrame
print("\nSparse PCA Components DataFrame:")
components_df.T
```

```python
import numpy as np
import pandas as pd
from sklearn.metrics.pairwise import rbf_kernel
from sklearn.cluster import DBSCAN

def cartesian_to_polar_nd(df, center=None):
    """
    Convert nD Cartesian coordinates to nD polar coordinates.
    
    Parameters:
        df (pd.DataFrame): DataFrame containing Cartesian coordinates with columns ["x1", "x2", ..., "xn"].
        center (array-like): Reference point (center for polar conversion). Defaults to the origin.
    
    Returns:
        pd.DataFrame: DataFrame with additional columns for r and theta_1, theta_2, ..., theta_(n-1).
    """
    # If no center is provided, assume the origin
    if center is None:
        center = np.zeros(df.shape[1])
    
    # Center the data
    centered_data = df.values - center
    
    # Compute radial distance (r)
    r = np.linalg.norm(centered_data, axis=1)
    
    # Compute angles (theta_1, theta_2, ..., theta_(n-1))
    thetas = []
    for i in range(df.shape[1] - 1):
        # Compute norm of the remaining dimensions
        norm_remaining = np.linalg.norm(centered_data[:, i+1:], axis=1)
        theta = np.arctan2(norm_remaining, centered_data[:, i])
        thetas.append(theta)
    
    # Create a new DataFrame for polar coordinates
    polar_df = pd.DataFrame()

    polar_df["r"] = r
    for i, theta in enumerate(thetas):
        polar_df[f"theta_{i+1}"] = theta
    
    return polar_df

# Step 1: Define the reference point (center for polar conversion)
reference_point = np.zeros(data.shape[1])  # np.zeros with length equal to the number of columns in df

# Step 2: Map the n-dimensional Cartesian space to n-dimensional polar space
polar_data = cartesian_to_polar_nd(data, reference_point)

# Step 3: Use only the polar variables (r and theta columns)
theta_columns = [col for col in polar_data.columns if col.startswith("theta")]
polar_features = ["r"] + theta_columns  # This includes only the new polar coordinates

# Step 4: Compute RBF kernel for the polar data
gamma = 1.0  # Hyperparameter to control the RBF spread
rbf_matrix = rbf_kernel(polar_data[polar_features].values, gamma=gamma)

# Step 5: Convert similarity matrix to distance matrix
distance_matrix = 1 - rbf_matrix

# Step 6: Apply DBSCAN with the precomputed distance matrix
epsilon = 0.1  # Set based on data's density
min_samples = 2
dbscan = DBSCAN(eps=epsilon, min_samples=min_samples, metric="precomputed")
labels = dbscan.fit_predict(distance_matrix)

# Add cluster labels to the DataFrame
polar_data["cluster"] = labels

# Output clustering results
print("Clustered Data (DataFrame):")
print(polar_data)
```
